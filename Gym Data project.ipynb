# Creating Data Table

```python
import datetime
import random

# Define start date and number of days
start_date = datetime.date(2023, 9, 26)
num_days = 100

# Initialize a list to store the generated dates
dates = []

# Generate dates
current_date = start_date
while len(dates) < num_days:
    # Add the current date to the list
    dates.append(current_date.strftime("%d/%m/%Y"))
    
    # Move to the next day
    current_date += datetime.timedelta(days=1)
    
    # Check if a day is missed randomly
    if random.random() < 3/7:  # Probability of missing a day (4 out of 7 days)
        # Check if the next day should also be missed
        if random.random() < 2/7:  # Probability of missing two consecutive days
            # Move to the day after the missed days
            current_date += datetime.timedelta(days=2)
            if random.random() < 1/7:  # Probability of missing two consecutive days
                # Move to the day after the missed days
                current_date += datetime.timedelta(days=3)
        else:
            # Move to the next day
            current_date += datetime.timedelta(days=1)
```

# Computing Moving Average for Weight @@@@@@@@@@@@@@@@

```python
weights = date_weight_imputed.body_weight  # Your weight measurements
window_size = 7  # Define the window size for the moving average (e.g., per week)

# Calculate the moving average of weight loss
moving_average_weight = []

for i in range(len(weights) - window_size + 1):
    # Calculate the average weight loss over the window
    average_loss = (weights[i + window_size - 1] - weights[i]) / window_size
    moving_average_weight.append(average_loss)

# Print the moving averages
#print("Moving averages of weight loss per day:")
#for i, average_loss in enumerate(moving_averages):
#  print(f"day {i + 1}: {average_loss:.2f} kg/day")
```
